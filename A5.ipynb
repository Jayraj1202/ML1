```python
import pandas as pd
import seaborn as sns 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
```

```python
x=data.iloc[:,3:].values
```

```python
x
```

```python
kmean=KMeans(n_clusters=4,random_state=0)
```

```python
kmean.fit(x)
```

```python
kmean.fit_predict(x)
```

```python
center=kmean.cluster_centers_
center
```

```python
kmean.inertia_
```

```python
labels=kmean.labels_
labels
```

```python
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=labels)
plt.scatter(center[:,0],center[:,1],color='red',marker='d',linewidth=6)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
wcss=[]
for i in range(1,10):
    kmean=KMeans(n_clusters=i)
    kmean.fit(x)
    wcss.append(kmean.inertia_)
    print("Inertias:",kmean.inertia_)

plt.plot(range(1,10),wcss,marker='o')
plt.title('Elbow')
plt.xlabel('No of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
km=KMeans(n_clusters=5,random_state=0)
km.fit(x)
```

```python
km.fit_predict(x)
```

```python
ctr=km.cluster_centers_
ctr
```

```python
km.inertia_
```

```python
labl=km.labels_
```

```python
labl
```

```python
data[labl==2]
```

```python
plt.scatter(data['Annual Income (k$)'],data['Spending Score (1-100)'],c=labl)
plt.scatter(ctr[:,0],ctr[:,1],c="red",marker='d',linewidth=7)
```

```python
fit=km.fit_predict(x)
plt.scatter(x[fit==0,0],x[fit==0,1],s=100,c='red',label='cluster 0')
plt.scatter(x[fit==1,0],x[fit==1,1],s=100,c='green',label='cluster 1')
plt.scatter(x[fit==2,0],x[fit==2,1],s=100,c='yellow',label='cluster 2')
plt.scatter(x[fit==3,0],x[fit==3,1],s=100,c='blue',label='cluster 3')
plt.scatter(x[fit==4,0],x[fit==4,1],s=100,c='orange',label='cluster 4')

plt.scatter(ctr[:,0],ctr[:,1],s=200,c='black',label='centroid',marker='*')
plt.legend()
plt.show()
```

```python
fit=km.fit_predict(x)
plt.scatter(x[fit==0,0],x[fit==0,1],s=100,c='red',label='c0')
plt.scatter(x[fit==1,0],x[fit==1,1],s=100,c='green',label='c1')
plt.scatter(x[fit==2,0],x[fit==2,1],s=100,c='yellow',label='c2')
plt.scatter(x[fit==3,0],x[fit==3,1],s=100,c='blue',label='c3')
plt.scatter(x[fit==4,0],x[fit==4,1],s=100,c='orange',label='c4')

plt.scatter(ctr[:,0],ctr[:,1],c='black',linewidth=6,marker='d',label='centroid')
new_data=[[15,30]]
predicted_cluster=km.predict(new_data)
plt.scatter(new_data[0][0],new_data[0][1],s=200,color='brown',marker='*',label='newdata')
plt.legend()
plt.show()
```
